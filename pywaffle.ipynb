```python
#!pip install pywaffle matplotlib
from pywaffle import Waffle
import matplotlib.pyplot as plt
```

```python
data = {'Rinat': 34, 'Liliya': 32, 'Nazar': 6, 'Elvina': 5}
fig = plt.figure(
    FigureClass=Waffle, 
    rows=5, 
    values=data, 
    colors=("#983D3D", "#232066", "#DCB732", "#052036" ),
    title={'label': 'Ismagilov\'s family ages at 19.03.2020', 'loc': 'left'},
    labels=["{0} ({1}%)".format(k, v) for k, v in data.items()],
    legend={'loc': 'lower left', 'bbox_to_anchor': (0, -0.4), 'ncol': len(data), 'framealpha': 0},
#     plot_direction='NW',
    tight=False,
    figsize=(9, 6)
)
fig.set_facecolor('#EEEEEE')
plt.show()
```
